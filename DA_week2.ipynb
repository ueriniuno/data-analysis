*이상치 정제하기

```python
 df = pd. DataFrame({ 'sex' : [1, 2, 1, 3, 2, 1 ], # 여기선 sex 1,2 중 하나 / score는 1-5
                     'score' : [5,4,3,4,2,6]})
 df
```

```python
df['sex'].value_counts().sort_index() #3이 이상치
```

```python
 df['score'].value_counts().sort_index() #6이 이상치
```

```python
 # Sexフト 3이면 NaN 부여, 아니면 원래 값 그대로 부여
df['sex'] = np.where(df['sex'] == 3, np.nan, df['sex'])
df
```

```python
 # SCOre가 5보다 크면 NaN 부勾
df['score'] = np.where(df['score'] > 5, np.nan, df['score'])
df
```

```python
df.dropna(subset = ['sex', 'score']) \
  .groupby('sex') \
    .agg(mean_score = ('score', 'mean'))
#sex, score의 nan값 제거
#sex별로 묶기  1,2
#sex1,2에서 각각 score의 평균값
```

1사분위수 .quantile(.25)
3사분위수  ~(.75)
- 이 둘의 거리 차를 iqr이라고 함.

```python
 pct25 = mpg[ 'hwy'].quantile(.25)
 pct25
```

```python
pct75 = mpg['hwy'].quantile(.75)
pct75
```

```python
iqr = pct75 - pct25
iqr
```

```python
 pct25 -1.5* iqr # 하한
```

```python
 pct75 + 1.5* iqr # 상한
```
