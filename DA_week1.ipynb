4) 모듈
- 함수가 많은 패키지는 비슷한 함수끼리 몇 개의 모듈로 나뉘어 있음. 패키지라는 큰 꾸러미에 비슷한 함수들끼리 모아둔 작은 꾸러미라고 생각.
패키지 -> 모듈 -> 함수

ex) 머신러닝 모델 만들 때 사용하는 sklearn 패키지 안에는 metrics, tree, model_selection 등 여러 모듈이 있다. metrics 모듈에는 머신러닝 모델의 예측이 얼마나 정확한지 성능을 평가할 때 사용하는 accuracy_score() 함수가 있다.
accuracy_score()를 사용하려면 먼저 import sklearn.metrics로 패키지를 로드하는 코드 뒤에 점을 찍고 모듈 이름을 입력한다.


import 패키지.모듈  # 패키지의 모듈 로드

패키지.모듈명.함수명()  #패키지, 모듈의 함수 사용

````python
import sklearn.metrics
sklearn.metrics.accuracy_score()  # 함수에 아무것도 입력하지 않아서 오류나는 거 당연


# 위의 두 번째 줄 처럼 패키지.모듈.함수 입력 번거로우면
# from 패키지명.모듈명 import 함수명 으로 로드해서 함수만 사용 가능
from sklearn import metrics
metrics.accuracy_score()

from sklearn.metrics import accuracy_score
accuracy_score()
````
