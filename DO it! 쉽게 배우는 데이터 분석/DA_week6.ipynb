```python
import pandas as pd
df_raw = pd.read_csv('exam.csv')
df = df_raw.head(10)
df
```

```python
# nclass 추출
df['nclass']
```

```python
# nclass가 1인지 확인
df['nclass'] == 1
```

```python
#nclass가 1이면 추출
df[df['nclass'] == 1]
```

```python
# 수학 점수가 80점 이상이면 추출
df[df['math'] >= 80]
```

```python
# nclass가 1이면서 수학 점수가 50점 이상
df[(df['nclass'] == 1) & (df['math'] >= 50)]
```

```python
#수학 점수가 50점 미만이거나 영어 점수가 50점 미만
df[(df['math'] < 50) | (df['english'] < 50)]
```

```python
exam = pd.read_excel('excel_exam.xlsx')
mpg = pd.read_csv('mpg.csv')

# df.query()로 목록에 해당하는 행 추출하기
exam.query('nclass in [1, 3, 5]')

# df.groupby()에 여러 변수 지정하기
mpg.groupby(['manufacturer', 'drv'])\
    .agg(mean_cty=('cty', 'mean'))
```
