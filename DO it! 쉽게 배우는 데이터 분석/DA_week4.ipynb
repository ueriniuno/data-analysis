##11-1. 시군구별 인구 단계 구분도 만들기

```python
import json
geo = json.load(open('SIG.geojson', encoding = 'UTF-8'))
```

```python
#행정 구역 코드 출력
geo['features'][0]['properties']
```

```python
#위도, 경도 좌표 출력
geo['features'][0]['geometry']
```

```python
#시군구별 인구 데이터 준비하기
import pandas as pd
df_pop = pd.read_csv('Population_SIG.csv')
df_pop.head()
```

```python
df_pop.info()
```

```python
df_pop['code'] = df_pop['code'].astype('str')
```

```python
pip install folium
```

```python
#배경지도 만들기 -> 지도 중심 좌표, 확대 단계
import folium
folium.Map(location = [35.95, 127.7], zoom_start = 8)
```

```python
map_sig = folium.Map(location = [35.95, 127.7], zoom_start = 8,
                     titles = 'cartodbpositron')
map_sig
```

```python
#folium.Choropleth()에 .add로 앞에서 만든 배경 지도 map_sig에 단계 구분도 씌움
folium.Choropleth(
    geo_data = geo,
    data = df_pop,
    columns = ['code', 'pop'],
    key_on = 'feature.properties.SIG_CD') \
    .add_to(map_sig)
map_sig
```

```python
#계급 구간 정하기(지역을 인구에 따라 다섯 단계의 색깔로 표현)
bins = list(df_pop['pop'].quantile([0, 0.2, 0.4, 0.6, 0.8, 1]))
bins
```

```python
#디자인 수정하기
map_sig = folium.Map(location = [35.95, 127.7], zoom_start = 8,
                     titles = 'cartodbpositron')
```

```python
folium.Choropleth(
    geo_data = geo,
    data = df_pop,
    columns = ['code', 'pop'],
    key_on = 'feature.properties.SIG_CD',
    fill_color = 'YlOrRd',
    fill_opacity = 1,
    line_opacity = 0.5,
    bins = bins ) \
    .add_to(map_sig)
map_sig
```

##11-2. 서울시 동별 외국인 인구 단계 구분도 만들기

```python
#서울시 동 경계 지도 데이터 준비
import json
geo_seoul = json.load(open('EMD_Seoul.geojson', encoding = 'UTF-8'))
```

```python
geo_seoul['features'][0]['properties']
```

```python
#위도 경도 좌표 출력
geo_seoul['features'][0]['geometry']
```

```python
#서울시 동별 외국인 인구 데이터 준비
foreigner = pd.read_csv('Foreigner_EMD_Seoul.csv')
foreigner.head()
```

```python
foreigner.info()
```

```python
foreigner['code'] = foreigner['code'].astype('str')
```

```python
#단계 구분도 만들기-> 지역을 8단계로 나눔
bins = list(foreigner['pop'].quantile([0, 0.2, 0.4, 0.6, 0.8, 1]))
bins
```

```python
#배경지도 만들기
map_seoul = folium.Map(location = [37.55, 126.98], zoom_start = 12,
                      titles = 'cartodbpositron')
```

```python
#단계구분도 만들기
folium.Choropleth(
    geo_data = geo_seoul,
    data = foreigner,
    columns = ('code', 'pop'),
    key_on = 'feature.properties.ADM_DR_CD',
    fill_color = 'Blues',
    nan_fill_color='White',
    fill_opacity = 1,
    line_opacity = 0.5,
    bins = bins) \
    .add_to(map_seoul)
map_seoul
```

```python
#구 경계선 추가하기
geo_seoul_sig = json.load(open('SIG_Seoul.geojson', encoding = 'UTF-8'))
```

```python
#서울 구 라인 추가
folium.Choropleth(
    geo_data = geo_seoul_sig,
    fill_opacity=0,
    line_weight = 4) \
    .add_to(map_seoul)
map_seoul
```

##12-1. 인터랙티브 그래프 만들기


```python
import pandas as pd
mpg = pd.read_csv('mpg.csv')
```

```python
#산점도 만들기
import plotly.express as px
px.scatter(data_frame=mpg, x = 'cty', y = 'hwy', color = 'drv')
```

```python
#자동차 종류별 빈도 구하기
df = mpg.groupby('category', as_index = False) \
        .agg(n = ('category', 'count'))
df
```

```python
#막대 그래프 만들기
px.bar(data_frame = df, x = 'category', y = 'n', color = 'category')
```

```python
#선그래프 만들기
economics = pd.read_csv('economics.csv')
px.line(data_frame = economics, x = 'date', y = 'psavert')
```

```python
#상자 그림 만들기
px.box(data_frame = mpg, x = 'drv', y = 'hwy', color = 'drv')
```

```python
#그래프를 변수에 할당하기
fig = px.scatter(data_frame=mpg, x = 'cty', y = 'hwy')
#html로 저장하기
fig.write_html('scatter.html')
```
